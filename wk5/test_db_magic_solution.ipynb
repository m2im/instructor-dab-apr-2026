# 🪄 Introduction to Magic Commands in Jupyter

**Magic commands** (or **IPython magics**) are special shorthand shortcuts built into the Jupyter notebook kernel that extend its core capabilities. They allow you to control your notebook's environment, interact with your computer's operating system, and run code in other languages (like SQL or Bash) without leaving your Python environment.

Magic commands always begin with a percent sign (`%`) and are divided into two main categories:

---

### 1. Line Magics (`%`)
Line magics operate on **a single line of code only**. Anything else typed on the lines below them will be treated as standard Python.

* `%load_ext sql`: Loads an external extension (like the SQL toolset).
* `%pip install pandas`: Installs a library to your current notebook environment without needing to open a separate terminal.
* `%pwd`: Prints the current working directory path of your notebook.

---

### 2. Cell Magics (`%%`)
Cell magics operate on the **entire notebook cell**. The magic command **must** be placed at the absolute top of the cell (Line 1); otherwise, Jupyter will throw a syntax error.

* `%%sql`: Converts the entire cell into an active SQL editor window, allowing you to run database queries directly against a connected database file.
* `%%timeit`: Runs the code inside the cell multiple times and calculates the exact average execution speed (ideal for testing data processing performance).
* `%%bash`: Runs script commands directly inside your operating system's terminal shell wrapper.

---

### 🎯 Why They Are Crucial for Data Analytics

Without magic commands, connecting a Jupyter notebook to a database engine like SQLite requires writing repetitive Python boilerplate code (`sqlite3.connect`, creating cursors, and wrapping queries inside text strings).

By using the combination of `%load_ext sql` and `%%sql`, you turn your data notebook into a **direct database terminal console**, letting you quickly prototype schemas and test queries on the fly.

**Note on Syntax:** The three slashes (///) indicate a relative path. This explicitly tells Jupyter to create the database file in the exact same folder directory where your current notebook file (.ipynb) is running.

### Create the table AUTHORS

```python
%%sql
CREATE TABLE authors (
    author_id INTEGER PRIMARY KEY AUTOINCREMENT,
    first_name TEXT NOT NULL,
    last_name TEXT NOT NULL,
    biography TEXT
);
```

### Create the table BOOKS

```python
%%sql
CREATE TABLE books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    isbn TEXT NOT NULL UNIQUE CHECK(length(isbn) BETWEEN 10 AND 13),
    price REAL NOT NULL CHECK(price >= 0.0),
    stock_quantity INTEGER NOT NULL DEFAULT 0 CHECK(stock_quantity >= 0)
);
```
