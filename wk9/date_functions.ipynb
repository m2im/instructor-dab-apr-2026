The output is the same because, while you provided different input formats for Pandas to interpret, the resulting data type in memory is identical.

Pandas does not "remember" the format you used to create the date; it only cares about the absolute point in time that the date represents. Once Pandas successfully parses your input, it standardizes the result into a datetime64 object, which is always displayed using a standard ISO-like format: ```YYYY-MM-DD HH:MM:SS```.

**Why this happens**

- **Parsing vs. Storing:** The format argument is a "map" that tells Pandas how to navigate your input string to find the year, month, and day. Once it finds those components, it discards the original string and stores the date internally as a numeric timestamp (the number of nanoseconds since January 1, 1970).

- **Standardized Display:** When you print() or display a datetime object in Jupyter, Pandas uses a default representation. It doesn't matter if you typed the date as "September 30th" or "09/30/23"—internally, they are the exact same point in time, so they are printed the same way.

**How to see them differently**

If you want to "see" your dates in different formats, you are no longer talking about the datetime64 object itself, but rather a string representation of that date. To achieve this, you use the ```.strftime()``` method.

Think of the input format as the language the data is written in, and the datetime object as the universal meaning behind those words. Pandas translates all those different languages into one universal, high-performance internal format so that it can perform calculations, sorting, and grouping on them efficiently.

### Working with Periods
A Period is fundamentally different from a Timestamp:

- **Timestamp:** Represents a single, specific point in time (e.g., "2023-01-01 09:00:00").

- **Period:** Represents a fixed span of time (e.g., the entire year of 2023).

**Why this is useful**

For some projects, periods are extremely powerful for downsampling and financial reporting. For example, if we have daily patrol logs but need to report budget or resource consumption by year or fiscal quarter, converting our data to Period objects allows us to group and aggregate the data cleanly:

```python
# Convert daily data to year-period
df['year'] = df['order_date'].dt.to_period('Y')

# Group by the period
annual_report = df.groupby('year')['total_price'].sum()

````python
# Create Period objects
# When we execute pd.Period('2023', freq='Y'), Pandas interprets this as a single, discrete Period representing the entirety of the year 2023.
# The Output Explained: Period('2023', 'Y-DEC')
# '2023': This is the label of the period. Since we specified the frequency as 'Y' (Year), Pandas sets the period to encompass the full chronological span of that year.
#'Y-DEC': This part is the Frequency Alias.
# Y stands for Year.
# -DEC tells us the "anchor" month. By default, Pandas assumes a standard calendar year, which ends in December. If we were working with a fiscal year that ended in June, you would see 'Y-JUN'.
````

# Difference between pandas.Timestamp and datetime64[us]
The confusion between pandas.Timestamp and datetime64[us] is common, as they both represent a specific point in time. The difference lies in whether you are looking at the object itself or the data type stored in a column.

**1.  pandas.Timestamp (The Object)**
Think of Timestamp as the individual unit of time. It is a single object that acts like a Python datetime object but with extra "superpowers" specifically for data science.

- **Usage:** You use it to represent one single moment in your code.

- **Capabilities:** Because it is a "Pandas" object, it has built-in methods for time-zone awareness, frequency handling, and arithmetic (e.g., ts + pd.Timedelta(days=1)).

**2. datetime64[us] (The Data Type)**
datetime64[us] is the memory representation (dtype) used by Pandas when it stores a column of dates.

- **Usage:** You see this when you inspect df.info() or df.dtypes. It describes how the data is packed into your computer's RAM.

- **The [us] part:** This specifically denotes microsecond precision. Other common types are datetime64[ns] (nanoseconds), which is the default for most modern Pandas operations.

- **Efficiency:** This is a NumPy-based type. It is optimized for performance, allowing Pandas to store millions of dates in a contiguous block of memory, which is much faster than storing millions of individual Timestamp objects.
