# Analyzing Sales Data

Let's calculate the monthly sales growth rate from our dataset.

The formula we'll use is:
$\text{Growth Rate} = \frac{\text{Current Month} - \text{Previous Month}}{\text{Previous Month}} \times 100\%$

First, we'll import the necessary libraries and calculate the month-over-month growth rates:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and process sales data
sales_data = pd.read_csv('monthly_sales.csv')
sales_data['Growth_Rate'] = sales_data['Sales'].pct_change() * 100

# Plot growth rates
plt.bar(sales_data['Month'], sales_data['Growth_Rate'].fillna(0))
plt.title('Monthly Sales Growth Rate')
plt.ylabel('Growth Rate (%)')
plt.show()
```
